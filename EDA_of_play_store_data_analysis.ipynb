# Play Store Data Analysis

This project involves analyzing the Play Store data to gain insights into the various aspects of mobile applications available on the platform. The dataset used in this analysis contains information about app categories, ratings, reviews, installs, prices, and more.

## Overview

The analysis covers the following aspects:

- Exploratory Data Analysis (EDA): Examining the distribution and characteristics of the Play Store data.
- Visualization: Creating visual representations such as bar charts, pie charts, and scatter plots to illustrate trends and patterns in the data.
- Insights: Drawing conclusions and extracting meaningful insights from the data analysis.

## Dataset

The  2 dataset used for this analysis is sourced from Almabetter. It consists of Many rows and Many columns, including information about app names, categories, ratings, reviews, installs, prices, sentiments,sentiment polarity and more.

## Analysis

### Exploratory Data Analysis (EDA)

The EDA phase involves:

- Data Cleaning: Handling missing values, removing duplicates, and formatting columns.
- Descriptive Statistics: Calculating summary statistics such as mean, median, and mode for numerical variables.
- Distribution Analysis: Examining the distribution of ratings, reviews, and installs across different categories.

### Visualization

Visualization is used to:

- Visualize Distribution: Creating histograms, box plots, and violin plots to visualize the distribution of numerical variables.
- Compare Categories: Generating bar charts and pie charts to compare different categories based on ratings, installs, and prices.
- Identify Patterns: Using scatter plots and pair plots to identify correlations and patterns between numerical variables.

### Insights

The insights derived from the analysis include:

- Popular Categories: Identifying the most popular app categories based on the number of installs and user ratings.
- Pricing Strategy: Analyzing the relationship between app prices and user ratings to formulate effective pricing strategies.
- User Engagement: Understanding factors influencing user engagement by examining app reviews and sentiment analysis.

## Conclusion

The analysis provides valuable insights into the Play Store ecosystem, helping app developers, marketers, and business owners make informed decisions regarding app development, pricing, and user engagement strategies.

# How has the number of app installs changed over time, and are there any seasonal trends in app downloads?



```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Convert the 'Date' column to datetime format
merge_data['Last Updated'] = pd.to_datetime(merge_data['Last Updated'])

# Set the 'Date' column as the index
merge_data.set_index('Last Updated', inplace=True)

# Resample the data to aggregate app installs by month
monthly_installs = merge_data.resample('M').sum()

# Plot the time series of app installs
plt.figure(figsize=(12, 6))
plt.plot(monthly_installs.index, monthly_installs['count_app_install'], marker='o', linestyle='-')
plt.title('Monthly App Installs Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Installs')
plt.grid(True)
plt.show()

# Decompose the time series into trend, seasonal, and residual components
decomposition = seasonal_decompose(monthly_installs['count_app_install'], model='additive')

# Plot the decomposed components
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.suptitle('Seasonal Decomposition of Monthly App Installs')
plt.show()

# Visualize seasonal patterns in app downloads using a heatmap
monthly_installs['Month'] = monthly_installs.index.month
monthly_installs['Year'] = monthly_installs.index.year
pivot_table = monthly_installs.pivot_table(index='Month', columns='Year', values='count_app_install', aggfunc='sum')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Monthly App Installs Over Years')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

# Can sentiment analysis of app reviews provide insights into user satisfaction and preferences within different categories?

```python
from textblob import TextBlob

# Perform sentiment analysis using TextBlob
merge_data['Sentiment'] = merge_data['Translated_Review'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Define thresholds for sentiment categories
def categorize_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Categorize sentiment into Positive, Negative, or Neutral
merge_data['Sentiment Category'] = merge_data['Sentiment'].apply(categorize_sentiment)

# Group data by category and sentiment category
sentiment_by_category = merge_data.groupby(['Category', 'Sentiment Category']).size().unstack(fill_value=0)

# Plot the stacked bar chart
plt.figure(figsize=(12, 8))
sentiment_by_category.plot(kind='bar', stacked=True, cmap='coolwarm')
plt.title('Sentiment Analysis of App Reviews by Category')
plt.xlabel('Category')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```

## What are the most common genres within each category, and do certain genres tend to have higher ratings or more installs?



# To answer this question, we can perform the following steps:


1.Group the data by category and genre to find the most common genres
within each category.

2.Calculate the average ratings and total installs for each genre.

3.Analyze if certain genres tend to have higher ratings or more installs.

```python
# Group data by category and genre to find the most common genres within each category
genre_counts = merge_data.groupby(['Category', 'Genres']).size().reset_index(name='Count')

# Sort the genres within each category by count in descending order
genre_counts_sorted = genre_counts.sort_values(by=['Category', 'Count'], ascending=[True, False])

# Display the most common genres within each category
print("Most common genres within each category:")
print(genre_counts_sorted.groupby('Category').head(5))
print()

# Calculate the average ratings and total installs for each genre
genre_stats = merge_data.groupby('Genres').agg({'Rating': 'mean', 'count_app_install': 'sum'}).reset_index()

# Sort genres by average rating in descending order
genre_stats_sorted_by_rating = genre_stats.sort_values(by='Rating', ascending=False)

# Sort genres by total installs in descending order
genre_stats_sorted_by_installs = genre_stats.sort_values(by='count_app_install', ascending=False)

# Display the genres with highest average ratings
print("Genres with highest average ratings:")
print(genre_stats_sorted_by_rating.head(5))
print()

# Display the genres with highest total installs
print("Genres with highest total installs:")
print(genre_stats_sorted_by_installs.head(5))
```
