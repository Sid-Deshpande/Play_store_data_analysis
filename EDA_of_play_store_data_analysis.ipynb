We divide the rating column into 4 categories

*  if rating is between 4-5 then it is top rated
*  if rating is between 3-4 then above average
*  if rating is between 2-3 then it is average
*  if rating is between 1-2 then it is below average
*  if rating is between 0-1 then it is poor





```python
# Define the rating categories and count the number of apps in each category
categories = ['Top Rated', 'Above Average', 'Average', 'Below Average', 'Poor']
counts = [
    ((play_store_data['Rating'] >= 4) & (play_store_data['Rating'] <= 5)).sum(),
    ((play_store_data['Rating'] >= 3) & (play_store_data['Rating'] < 4)).sum(),
    ((play_store_data['Rating'] >= 2) & (play_store_data['Rating'] < 3)).sum(),
    ((play_store_data['Rating'] >= 1) & (play_store_data['Rating'] < 2)).sum(),
    (play_store_data['Rating'] < 1).sum()
]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=categories, autopct='%1.1f%%', startangle=150)
plt.title('Distribution of Ratings Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
